```python
import requests
import mysql.connector
import datetime

# Set up the API parameters
api_key = "YOUR API KEY"
api_base_url = "http://api.weatherstack.com/current"
cities = ["Kolkata", "Mumbai", "Bangalore", "Delhi", "Karachi", "Taipei", "London", "Bristol"]

# Set up the database connection
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="YOUR PASSWORD",
    database="YOUR DATABASE NAME",
    port=3306
)

# Define a function to insert weather data into the database
def insert_weather_data(city_name, temperature, pressure, humidity):
    now = datetime.datetime.now()
    date_time = now.strftime('%Y-%m-%d %H:%M:%S')
    cursor = db.cursor()
    sql = "INSERT INTO weather_data (city_name, temp, pressure, humidity, date_time) VALUES (%s, %s, %s, %s, %s)"
    val = (city_name, temperature, pressure, humidity, date_time)
    cursor.execute(sql, val)
    db.commit()
    print(cursor.rowcount, "record inserted.")

# Fetch weather data for each city and insert it into the database
for city in cities:
    url = f"{api_base_url}?access_key={api_key}&query={city}"
    response = requests.get(url)
    data = response.json()
    if 'current' not in data:
        print(f"Error fetching data for {city}: 'current' key not found in response")
        continue
    temperature = data['current']['temperature']
    pressure = data['current']['pressure']
    humidity = data['current']['humidity']
    insert_weather_data(city, temperature, pressure, humidity)

# Close the database connection
db.close()
```

```python
import mysql.connector

# Set up the database connection
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="YOUR PASSWORD",
    database="YOUR DATABASE NAME",
    port=3306
)

# Create a cursor object to interact with the database
cursor = db.cursor()

# Define a SQL query to retrieve the data from the table
query = "SELECT * FROM weather_data"

# Execute the query
cursor.execute(query)

# Fetch all the rows
records = cursor.fetchall()

# Display the data
if records:
    print("Weather Data:")
    for record in records:
        print("City:", record[1])
        print("Temperature:", record[2])
        print("Pressure:", record[3])
        print("Humidity:", record[4])
        print("Date and Time:", record[5])
        print("-------------------------")
else:
    print("No records found in the database.")

# Close the cursor and database connection
cursor.close()
db.close()
```
